```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
# import sys
# sys.path.insert(0, 'REPLACE_WITH_PWD')

import os
os.getcwd() # set current pwd as working directory

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('result_kMeans_with_label_attached.tsv', sep='\t', encoding='utf-8')
```

```python
df.info()
```

```python
from math import log
```

```python
df['dau_log'] = df['dau'].apply(lambda x: log(x, 10))
df['estimated_revenue_usd_log'] = df['estimated_revenue_usd'].apply(lambda x: log(x, 10))
```

```python
sb.lmplot('dau', 'estimated_revenue_usd'
         , data=df
         , fit_reg=False
         , hue = 'topics'
         , scatter_kws={'marker': 'D', 's': 100}
         , size=50)
plt.title('What users talk about?')
plt.xlabel('DAU')
plt.ylabel('Estimated Revenue Made$')
```

```python
sb.lmplot('dau_log', 'estimated_revenue_usd_log'
         , data=df
         , fit_reg=False
         , hue = 'topics'
         , scatter_kws={'marker': 'D', 's': 100}
         , size=20)
plt.title('What users talk about?')
plt.xlabel('DAU')
plt.ylabel('Estimated Revenue Made$')
```
